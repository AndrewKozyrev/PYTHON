```python
from IPython.display import clear_output
import random
from time import sleep

def display_board(board):
    clear_output()
    s = '|'
    print( " {0:-^32} ".format('') )
    for i in range(3):
        print( 3*(s+10*' ')+s)
        print( 3*(s+10*' ')+s)
        print( "| {0:^8} | {1:^8} | {2:^8} |".format( board[7-3*i], board[8-3*i], board[9-3*i] ) )
        print( 3*(s+10*' ')+s)
        print( 3*(s+10*' ')+s)
        print( " {0:-^32} ".format('') )
        
def player_input():
    '''
    OUTPUT = (player1_marker, player2_marker)
    '''
    marker = ''
    while marker != 'X' and marker != 'O':
        marker = input("Player 1: Do you want to be X or O?\n").upper()
    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')

def place_marker(board, marker, position):
    board[position] = marker
    
def win_check(board, mark):
    flagged = []
    for i in range(1, 10):
        if board[i] == mark:
            flagged.append(i) 
    win_patterns = [[1, 2, 3], [1, 4, 7], [2, 5, 8], [4, 5, 6], [3, 6, 9], [7, 8, 9], [1, 5, 9], [3, 5, 7]]
    for element in win_patterns:
        if set(element) <= set(flagged):
            return True
    return False

def choose_first():
    return ("player" + str(random.randint(1, 2)) )

def space_check(board, position):
    return not (board[position] == 'X' or board[position] == 'O')

def full_board_check(board):
    for i in range(1, 10):
        if space_check(board, i):
            return False
    return True

def player_choice(board):
    position = 0
    while not 1 <= position <= 9 or not space_check(board, position):
        position = int(input("Place your next marker:"))
    return position

def replay():
    sleep(3)
    ans = input("Do you want to play again?\n")
    return ans.lower() in ["yes", "i do", "ofcourse", "why not", "sure", "yep", "hell yeah!"]

print('Welcome to Tic Tac Toe!')

while True:
    b = ['#','','','','','','','','','']
    # Set the game up here
    player1_marker, player2_marker = player_input()
    players = {'player1': player1_marker, 'player2': player2_marker}
    fp = choose_first()
    sp = str('player1' if fp == 'player2' else 'player2')
    display_board(b)
    while True:
        #Player 1 Turn
        print(fp.capitalize() + " turn.\n")
        position = player_choice(b)
        place_marker(b, players[fp], position)
        display_board(b)
        if (win_check(b, players[fp])):
            print("Congradulations! " + fp.capitalize() + " have won the game!")
            break
        elif (full_board_check(b)):
            print("Tie!")
            break
        print(sp.capitalize() + " turn.\n")
        position = player_choice(b)
        place_marker(b, players[sp], position)
        display_board(b)
        if (win_check(b, players[sp])):
            print("Congradulations! " + sp.capitalize() + " have won the game!")
            break
    if not replay():
        break
```
